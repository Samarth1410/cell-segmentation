```python
import kagglehub

print("[INFO] Downloading dataset...")
path = kagglehub.dataset_download("aayushkt/cell-segmentation")
print(f"[INFO] Downloaded to: {path}")
```

```python
for root, dirs, files in os.walk(path):
    level = root.replace(path, "").count(os.sep)
    indent = "  " * level
    print(f"{indent}{os.path.basename(root)}/")
    if level < 3:
        for f in sorted(files)[:3]:
            print(f"{indent}  {f}")
        if len(files) > 3:
            print(f"{indent}  ... ({len(files)} files total)")
```

```python
import shutil

TARGET_DIRS = [
    "cell_dataset/train/train_images",
    "cell_dataset/train/train_masks",
    "cell_dataset/test/test_images",
]
for d in TARGET_DIRS:
    os.makedirs(d, exist_ok=True)

FOLDER_MAP = {
    "train_images": "cell_dataset/train/train_images",
    "train_masks":  "cell_dataset/train/train_masks",
    "test_images":  "cell_dataset/test/test_images",
}

def copy_files(src, dst):
    imgs = [f for f in os.listdir(src)
            if f.lower().endswith((".png", ".jpg", ".jpeg", ".tif"))]
    for f in imgs:
        shutil.copy2(os.path.join(src, f), os.path.join(dst, f))
    print(f"  ✓ {len(imgs)} files: {src} → {dst}")

matched = set()
for root, dirs, files in os.walk(path):
    folder_name = os.path.basename(root).lower().strip()
    for key, target in FOLDER_MAP.items():
        if key in folder_name and root not in matched:
            matched.add(root)
            copy_files(root, target)
            break
```

```python
for d in TARGET_DIRS:
    count = len(os.listdir(d))
    status = "✓" if count > 0 else "✗ EMPTY"
    print(f"  {status}  {d}/ ({count} files)")
```
